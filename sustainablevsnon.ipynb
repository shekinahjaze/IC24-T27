```python
import matplotlib.pyplot as plt
import numpy as np

sustainable = np.array([4.5,3.95,4.25,3.99,4.11,4.07,3.84,4.18,4.1,4.07,3.84,4.16,4.05,3.79,3.81])
nonsus = np.array([4.22,3.52,4.19,3.61,3.99,4.06,3.86,4.05,3.99,4.04,3.84,4.08,3.59,3.78,3.75])
plt.plot(sustainable, color='purple')
plt.plot(nonsus, color='yellow')
plt.xlabel('Question')
plt.ylabel('Perceived QOL')
plt.title('Comparing the Mean QOL Between a Sustainable and Non-Sustainable University ')
legend = plt.legend(('sustainable','non-sustainable'),loc='upper right')
plt.rcParams.update({
    "figure.facecolor":  (0.0, 0.0, 0.0, 0.0),  
    "axes.facecolor":    (0.0, .0, 0.0, 0.0),  
    "savefig.facecolor": (0.0, 0.0, 0.0, 0.0),  
})
legend.get_frame().set_alpha(None)
legend.get_frame().set_facecolor((0, 0, 1, 0.1))
plt.show()
plt.savefig('qol.png',transparent=True)
```
